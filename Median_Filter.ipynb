```python
import numpy as np
import matplotlib.pyplot as plt

# Function to apply a median filter to an image
def median_filter(image, filter_size):
    """
    Apply a median filter to a 2D image.
    :param image: 2D numpy array representing the grayscale image
    :param filter_size: Size of the filter (must be odd)
    :return: Filtered image as a 2D numpy array
    """
    # Ensure the filter size is odd
    if filter_size % 2 == 0:
        raise ValueError("Filter size must be odd.")

    # Calculate padding size
    pad_size = filter_size // 2

    # Pad the image with zeros
    padded_image = np.pad(image, pad_size, mode='constant', constant_values=0)

    # Initialize the output image
    filtered_image = np.zeros_like(image)

    # Traverse through the image
    for i in range(image.shape[0]):
        for j in range(image.shape[1]):
            # Extract the region of interest
            roi = padded_image[i:i + filter_size, j:j + filter_size]
            # Compute the median and assign it to the filtered image
            filtered_image[i, j] = np.median(roi)

    return filtered_image


# Example input image (5x5 matrix)
input_image = np.array([
    [10, 10, 10, 10, 10],
    [10, 50, 50, 50, 10],
    [10, 50, 100, 50, 10],
    [10, 50, 50, 50, 10],
    [10, 10, 10, 10, 10]
], dtype=np.uint8)

# Apply median filter
filter_size = 3
output_image = median_filter(input_image, filter_size)

# Display input and output images
plt.figure(figsize=(10, 5))
plt.subplot(1, 2, 1)
plt.title("Input Image")
plt.imshow(input_image, cmap='gray', vmin=0, vmax=255)
plt.colorbar()
plt.subplot(1, 2, 2)
plt.title("Filtered Image")
plt.imshow(output_image, cmap='gray', vmin=0, vmax=255)
plt.colorbar()
plt.tight_layout()
plt.show()
```
